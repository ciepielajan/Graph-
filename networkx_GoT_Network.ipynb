## Network Visualization in Python

- Helper notebook for article of same name published on Medium.

### Import

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
```

### Load and process data

```python
# load data
df = pd.read_csv("https://raw.githubusercontent.com/imohitmayank/got_network_viz_python/main/data/book1.csv")
# pick only important weights (hard threshold)
df = df.loc[df['weight']>10, :]
df
```

```python
# import
import networkx as nx
# load pandas df as networkx graph
G = nx.from_pandas_edgelist(df, 
                            source='Source', 
                            target='Target', 
                            edge_attr='weight')
print("No of unique characters:", len(G.nodes))
print("No of connections:", len(G.edges))
```

## Option 1: NetworkX

https://networkx.org/documentation/stable/reference/generated/networkx.drawing.nx_pylab.draw_networkx.html

```python
# all graph options
graphs_viz_options = [nx.draw, nx.draw_networkx, nx.draw_circular, nx.draw_kamada_kawai, nx.draw_random, nx.draw_shell, nx.draw_spring]

# plot graph option
selected_graph_option = 0

# plot
plt.figure(figsize=(15,10), dpi=50) 
graphs_viz_options[selected_graph_option](G, node_size=50)
```

```python
# rodzaje wykresów !!! 
graphs_viz_options = [nx.draw, 
                      nx.draw_networkx, 
                      nx.draw_circular, 
                      nx.draw_kamada_kawai, 
                      nx.draw_random, 
                      nx.draw_shell, 
                      nx.draw_spring] 

# wyświetlenie wszytkich możliwoych opcji

for i in range(len(graphs_viz_options)):
# plot graph option
  selected_graph_option = i

  # plot
  plt.figure(figsize=(8,6), dpi=100) 
  graphs_viz_options[selected_graph_option](G, node_size=50)
```
